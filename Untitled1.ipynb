```python
VOCALI=set('aeiouAEIOU')
```

```python
parola='mariavergine'
```

```python
def conta_vocali(parola):
    return sum(c in VOCALI for c in parola)
```

```python
conta_vocali('mariavergine')
```

```python
sorted(parole,
       key=lambda parola: sum (c not in VOCALI for c in parola)/len(parola),
       reverse=True
)[:10]
```
